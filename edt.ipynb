```python
import matplotlib.pyplot as plt
from scipy.ndimage.morphology import distance_transform_edt as edt
import numpy as np
from PIL import Image
```

```python
img = Image.open('./data/GlaS/testA_1_anno.bmp')
input_img = Image.open('./data/GlaS/testA_1.bmp')

plt.imshow(input_img)
```

```python
plt.imshow(img)
```

```python
np_img = np.array(img)
```

```python
binary_img = (np_img > 0.5) * 1.0
plt.imshow(binary_img)
```

```python
edt_img = edt(binary_img)
plt.imshow(edt_img)
```

```python
edt_inverse = edt((1 - binary_img))
plt.imshow(edt_inverse)
```
